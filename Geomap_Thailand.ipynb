```python
# import libraries
import pandas as pd
import folium
```

```python
# Load a data file
df = pd.read_csv("CovidAccumuate_Thailand.csv") 
```

```python
mp = folium.Map(location=[15.0000, 100.0000], zoom_start=5) # [latitude,longitude], zoom level = 5
```

```python
# Setup geomap
mp.choropleth(
    geo_data='thailand.json',
    key_on='feature.properties.name',
    data=df,
    columns=['province','total'],
    fill_color='YlOrRd',
    fill_opacity=2,
    line_opacity=0.2,
    legend_name='Province')
```
